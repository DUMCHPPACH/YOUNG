```python
t_list = [8]
x = 8
y = 15
for i in range(179):
    x += y
    t_list.append(x)

print(len(t_list))
```

```python
t_list2 = [12]
z = 12
y = 15
for i in range(179):
    z += y
    t_list2.append(z)

print(len(t_list2))
```

```python
all_line = []
with open('./label150 train_test.txt','rt', encoding='UTF8') as f:
    for i in range(t_list2[-1]):
        targetline = f.readline().rstrip()
        all_line.append(targetline)

print(len(all_line))
```

```python
train_acc = []
for i in t_list:
    acc = all_line[i-1].split('train acc ')[1][:9]
    train_acc.append(float(acc))
```

```python
print(len(train_acc))
print(train_acc)
```

```python
# test
```

```python
test_acc = []
for i in t_list2:
    testacc = all_line[i-1].split('test acc ')[1][:9]
    test_acc.append(float(testacc))

len(test_acc)
```

```python
fig, loss_ax = plt.subplots()
acc_ax = loss_ax.twinx()

loss_ax.plot(train_acc, 'y', label='train_acc')
loss_ax.plot(test_acc, 'r', label='test_acc')
loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('acc')
loss_ax.legend(loc='upper left')
```
