# 06. 비선형 활성화 함수(Activation function)

```python
# 그래프를 직접 그려보여 이해하자
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 시그모이드 함수(Sigmoid function)와 기울기 소실

def sigmoid(x):
    return 1/(1+np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.plot([0,0], [1.0, 0.0], ':')
plt.title('Sigmoid Function')
plt.show()
```

```python
# 하이퍼볼릭탄젠트 함수(Hyperbolic tangent function)

x = np.arange(-5.0, 5.0, 0.1)
y = np.tanh(x)

plt.plot(x, y)
plt.plot([0,0], [1.0, -1.0], ':')
plt.axhline(y=0, color='orange', linestyle='--')
plt.title('Tahn Function')
plt.show()
```

```python
# 렐루 함수(ReLU)

def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.plot([0,0],[5.0,0.0],':')
plt.title('Relu Function')
plt.show()
```

```python
# 리키 렐루(Leaky ReLU) - 죽은 랠루를 보완하기 위해 생김 leaky(새는) 정도를 조절 가능

# 기울기 조절 변수
a = 0.1

def leaky_relu(x):
    return np.maximum(a*x, x)

x = np.arange(-5.0, 5.0, 0.1)
y = leaky_relu(x)

plt.plot(x, y)
plt.plot([0,0], [5.0, 0.0], ':')
plt.title('Leaky Relu Function')
plt.show()
```
